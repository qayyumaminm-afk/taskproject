```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Create basic visualizations:
```

```python
sns.scatterplot(
    x='sepal length (cm)', 
    y='sepal width (cm)', 
    hue='target', 
    data=dataframe
)

plt.title("Sepal Length vs Sepal Width")
plt.xlabel("Sepal Length")
plt.ylabel("Sepal Width")
plt.show()
```

```python
# Histogram
plt.hist(dataframe['petal length (cm)'], bins=20)

plt.title("Distribution of Petal Length")
plt.xlabel("Petal Length")
plt.ylabel("Frequency")
plt.show()
```

```python
sns.boxplot(x='target', y='petal width (cm)', data=dataframe)

plt.title("Petal Width by Target")
plt.xlabel("Target")
plt.ylabel("Petal Width")
plt.show()
```
